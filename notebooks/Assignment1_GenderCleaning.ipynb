```python
# Import pandas
import pandas as pd

# Upload dataset
from google.colab import files
uploaded = files.upload()

# Read the dataset
df = pd.read_csv("EmergencyTriageDataset_Reduced_Dirty.csv")

# Display first 5 rows
print("First 5 rows of dataset:")
print(df.head())

# Check Gender column before cleaning
print("\nGender column before cleaning:")
print(df["Gender"].value_counts())

# Clean Gender column
df["Gender"] = df["Gender"].replace({
    "Male": 1,
    "MALE": 1,
    "male": 1,

    "Female": 0,
    "FEMALE": 0,
    "female": 0
})

# Convert column to integer
df["Gender"] = df["Gender"].astype(int)

# Display results after cleaning
print("\nGender column after cleaning:")
print(df["Gender"].value_counts())

# Show datatype
print("\nData type:")
print(df["Gender"].dtype)

# Display cleaned dataset
print("\nCleaned Dataset:")
print(df.head())

# Save cleaned dataset
df.to_csv("EmergencyTriageDataset_Cleaned.csv", index=False)

print("\nDataset cleaned and saved successfully!")
```
